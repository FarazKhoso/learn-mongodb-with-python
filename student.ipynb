```python
# ----------------------------
#  Connect to MongoDB Server
# ----------------------------

from pymongo import MongoClient

MONGODB_URI = "REPLACE_WITH_YOUR_MONGODB" # Replace with your mongodb uri

client = MongoClient(MONGODB_URI)  # Replace with your MongoDB URI
print("Connected to MongoDB server")
```

### Create or Select Database

```python
db = client["mydatabase"]  # Database name: mydatabase
print("Database selected:", db.name)
```

```python
client.list_database_names()
```

### Create or Select Collection

```python
collection = db["mycollection"] # Collection name: mycollection
print("Collection selected")
```

### Insert Documents

```python
# Insert single document
doc1 = {"name": "Faraz", "age": 19, "city": "Karachi"}
collection.insert_one(doc1)

# Insert multiple documents
docs = [
    {"name": "Ali", "age": 30, "city": "Lahore"},
    {"name": "Ahmed", "age": 28, "city": "Islamabad"}
]
collection.insert_many(docs)
print("Documents inserted")
```

### Find Documents

```python
# Find all documents
all_docs = collection.find({})
print("All documents:")
for doc in all_docs:
    print(doc)

# Find document with filter
faraz_doc = collection.find_one({"name": "Faraz"})
print("Document with name Faraz:", faraz_doc)

# Find with operators
age_gt_25 = collection.find({"age": {"$gt": 25}})
print("Documents with age > 25:")
for doc in age_gt_25:
    print(doc)
```

### Update Documents

```python
# Update a single document
collection.update_one(
    {"name": "Faraz"},  # Filter
    {"$set": {"age": 26}}  # Update operation
)

# Update multiple documents
collection.update_many(
    {"age": {"$lt": 30}},  # Filter
    {"$inc": {"age": 1}}    # Increment age by 1
)
print("Documents updated")

# Check updates
updated_docs = collection.find({})
print("Updated documents:")
for doc in updated_docs:
    print(doc)
```

### Delete Documents

```python
# Delete single document
collection.delete_one({"name": "Ahmed"})

# Delete multiple documents
collection.delete_many({"age": {"gt": 30}})  # Note: Adjust filter if needed
print("Documents deleted")

# Check remaining documents
remaining_docs = collection.find({})
print("Remaining documents:")
for doc in remaining_docs:
    print(doc)
```

### Query Operators Examples

```python
# $eq, $ne, $gt, $gte, $lt, $lte, $in, $nin, $exists, $regex
# Example: Find documents where age >= 26 and city exists
query_docs = collection.find({"age": {"$gte": 26}, "city": {"$exists": True}})
print("Documents with age >= 26 and city exists:")
for doc in query_docs:
    print(doc)
```
